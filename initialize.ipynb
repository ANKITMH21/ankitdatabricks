```python
%sql
SELECT *, row_number() OVER (ORDER BY Salary DESC) AS rank
FROM default.employee
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *
```

```python

df_rank = df.withColumn('rank',row_number().over(Window.orderBy(col('salary').desc())))
df_rank.display()
```

```python

df_rank = df.withColumn(
    "rank",
    row_number().over(Window.orderBy(col("salary").desc()))
)

display(df_rank)
```

Employee earning more than or equals to total city earning

```python
df_grp = df.groupBy(col('City')).agg(sum(col('Salary')).alias('Total'))

df_grp.display()
```

```python
df_check = df.join(df_grp,df.City==df_grp.City,'left_outer')\
    .filter(col('salary') >= col('Total'))
df_check.display()
```

```python
%sql

SELECT * 
FROM default.employee e 
LEFT JOIN (
    SELECT city,SUM(Salary) AS Total
    FROM default.employee
    GROUP BY city
)
t ON e.city = t.city
WHERE e.Salary >= t.Total
```
